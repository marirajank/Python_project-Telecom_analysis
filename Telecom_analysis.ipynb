#  2. Analyzing the 'state' vaiable

```python
data.head(2)
```

```python
data.state.nunique()               # To show the total number of unique values present in the state column 
```

```python
data.state.unique()                  # to show the unique values in state column
```

```python
data.state.value_counts(ascending = True)  
```

```python
#countplot

plt.figure(figsize = (20,7))

sns.set(style ='darkgrid')
b = sns.countplot(data.state)

plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.xlabel('State', fontsize ='15', color='green')
plt.ylabel('count', fontsize ='15', color='green')
plt.title('Customers by state', fontsize= 15)

for bars in b.containers:
    b.bar_label(bars)

plt.show()
```

```python
#showing churn/nont churn state-wise

plt.figure(figsize = (20,9))

sns.set(style ='whitegrid')
b = sns.countplot(data.state, hue = data.churn, palette=['#3FE517',"#FFCD02"] )

plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.xlabel('State', fontsize ='15', color='green')
plt.ylabel('count', fontsize ='15', color='green')
plt.title('Churned and Not-churned customers by state', fontsize= 15)

plt.legend(labels=['Not Churned', 'Churned'])
for bars in b.containers:
    b.bar_label(bars)

plt.show()
```

```python
# drawing a line chart for state-wise churn rate

x = data.state.unique()
y = data.groupby('state')['churn'].mean()

plt.figure(figsize = (25,8))
sns.set(style = 'darkgrid')

plt.plot(x,y, color = '#060560', marker = '.', markersize = 11)

plt.title('State Churn graph', fontsize = 20)
plt.xlabel('State', fontsize = 15 )
plt.xticks(fontsize= 12)

plt.show()

```

```python
#state churn percentage

c= pd.crosstab( data.state, data.churn, margins=True )

c['percentage_churn'] = c[True] / (c['All']) * 100           # creating new columnn 'percentage_churn'

print(c.head())
```

```python
c.info()
```

```python
# top 5 records according to percentage_churn

c.sort_values(by = 'percentage_churn', ascending = False).head()
```

#  9. Analyzing the 'per minute charge'

```python
data.head(3)
```

```python
# calls charges per minute

dc = data['total day charge'].mean() / data['total day minutes'].mean()
ec = data['total eve charge'].mean() / data['total eve minutes'].mean()
nc = data['total night charge'].mean() / data['total night minutes'].mean()
intc = data['total intl charge'].mean() / data['total intl minutes'].mean()
```

```python
print(dc)
print(ec)
print(nc)
print(intc)
```

```python
# drawing a bar plot to represent 'calls per min charges'

plt.figure(figsize=(8,5))
sns.set(style = 'darkgrid')
sns.barplot(x=['Day','Evening','Night','Internation'], y=[dc,ec,nc,intc], palette =['#01F50F','#FF7F0D','#025C83','#DD39E7'])

plt.show()
```
